```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Date": pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01"]),
    "Value":[10,15,22,28,35,39]
}
df= pd.DataFrame(data)
df.set_index("Date", inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index, df["Value"], marker='o',color="green")
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time series with Trend")
plt.grid(True)
plt.show()
```
