```python
import folium
mymap = folium.Map(location=[0, 0], zoom_start=2)
folium.Marker(location=[40.7128, -74.0059], popup='New York City', icon=folium.Icon(color='blue') ).add_to(mymap)
folium.Marker(location=[51.5074, -0.1276], popup='London', icon=folium.Icon(color='green')).add_to(mymap)
folium.Marker(location=[35.6895, 139.6917], popup='Tokyo', icon=folium.Icon(color='red')).add_to(mymap)

mymap
```

```python
import folium

cities = [
    {"name": "pu", "location": [13.1718, 77.5362], "symbol": "university"},
    {"name": "vignesh house", "location": [12.9767, 77.5713], "symbol": "hospital]"},
    {"name": "chinnaswamy stadium", "location": [12.9788, 77.5997], "symbol": "stadium"},

]

mymap = folium.Map(location=[0, 0], zoom_start=2)

for city in cities:
     folium.Marker(location=city["location"], popup=city["name"], icon=folium.Icon(icon=city["symbol"], prefix='fa')).add_to(mymap)
mymap

```

```python
import folium

cities = [
    {"name": "chinnaswami stadium", "location": [12.9788, 77.5997], "image": "/content/275679.webp"},
    {"name": "pu", "location": [13.1718, 77.5362], "image": "lon"},
    {"name": "vignesh house", "location": [12.9767, 77.5713], "image": "tokyo.jpg"},

]


mymap = folium.Map(location=[0, 0], zoom_start=2)


for city in cities:
    html = f'<img src="{city["image"]}" alt="{city["name"]}" width="200px">'
    folium.Marker(location=city["location"], popup=html).add_to(mymap)

mymap
```
