Vector Visualization


```python
import numpy as np
import matplotlib.pyplot as plt
x_pos=0
y_pos=0
a_pos=0
b_pos=0
x_direct=-1
y_direct=1
a_direct=1
b_direct=-1
fig,ax=plt.subplots(figsize=(12,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct)
ax.set_title('quiver plot with two arrow')
plt.show()
```

```python
x_pos=0
y_pos=0
a_pos=0
b_pos=0
x_direct=-1
y_direct=1
a_direct=1
b_direct=-1
fig,ax=plt.subplots(figsize=(20,20))
ax.quiver(x_pos,y_pos,x_direct,y_direct,color='r',scale=20)
ax.quiver(a_pos,b_pos,a_direct,b_direct,color='g',scale=10)
ax.set_title('quiver plot with two arrow')
plt.show()
```

```python

x = np.arange(-2, 2.2, 0.2)
y = np.arange(-2, 2.2, 0.2)

X, Y = np.meshgrid(x, y)
z = X * np.exp(-X**2-Y**2)
dx, dy = np.gradient(z)
fig, ax = plt.subplots(figsize =(9, 9))
ax.quiver(X, Y, dx, dy)
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')
plt.show()
```
