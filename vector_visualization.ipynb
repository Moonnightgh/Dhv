Vector Visualization


```python
import numpy as np
import matplotlib.pyplot as plt
x_pos=0
y_pos=0
a_pos=0
b_pos=0
x_direct=-1
y_direct=1
a_direct=1
b_direct=-1
fig,ax=plt.subplots(figsize=(12,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct)
ax.set_title('quiver plot with two arrow')
plt.show()
```

```python
x_pos=0
y_pos=0
a_pos=0
b_pos=0
x_direct=-1
y_direct=1
a_direct=1
b_direct=-1
fig,ax=plt.subplots(figsize=(12,7))
ax.quiver(x_pos,y_pos,x_direct,y_direct)
ax.quiver(a_pos,b_pos,a_direct,b_direct)
ax.set_title('quiver plot with two arrow')
plt.show()
```
